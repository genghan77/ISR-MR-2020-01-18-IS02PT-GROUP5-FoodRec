<h3>Functions</h3>

```python
def UpdateDataLabels(dataUpdate, foodGroups, foodSubGroups, labelNames, keyWords, value):
    if (type(foodGroups) is str):
        foodGroups = [foodGroups]
        
    if (type(foodSubGroups) is str):
        foodSubGroups = [foodSubGroups]
        
    if (type(labelNames) is str):
        labelNames = [labelNames]
        
    for group in foodGroups:
        #print(f'FoodGroup = {group}')
        if (foodSubGroups is None):
            foodSubGroups = dataUpdate[(dataUpdate['FoodGroup'] == group)]['FoodSubGroup'].unique()
        
        for subgroup in foodSubGroups:
            #print(f'FoodSubGroup = {subgroup}')
            
            for name in labelNames:
                #print(f'LabelName = {name}')
                
                for word in keyWords:
                    dataUpdate.loc[(dataUpdate['FoodGroup'] == group) & (dataUpdate['FoodSubGroup'] == subgroup) & \
                                   (dataUpdate['FoodName'].str.contains(word, case=False)), name] = value
                    
    listCounts=[]
    for name in labelNames:
        listCounts.append(dataUpdate[name].value_counts())

    display(pd.concat(listCounts, axis=1))
```

<h4>Summarise the Assam Prawn duplicates into small and large, with the respective nutrition values</h4>

```python
dictSmall={}
dictLarge={}

for col in data.columns:
    uniqueValues = data[data['FoodName'] == duplicated_names[0]][col].unique()
    
    if ((col not in num_group) & (col not in group)):
        dictSmall[col]=uniqueValues[0]
        dictLarge[col]=uniqueValues[0]
    elif (col not in num_group):
        print(uniqueValues)
        dictSmall[col]=uniqueValues[0]#.str.contains('16cm')
        dictLarge[col]=uniqueValues[1]#.str.contains('23cm')
        
    if (col in num_group):
        print(f'{col} : {uniqueValues}')
        dictSmall[col]=min(uniqueValues)
        dictLarge[col]=max(uniqueValues)

dictSmall['FoodName'] = dictSmall['FoodName'] + ' (small)'
dictLarge['FoodName'] = dictLarge['FoodName'] + ' (large)'

display(dictSmall)
display(dictLarge)
```

<h4>Update data</h4>

```python
data = data[data['FoodName'] != duplicated_names[0]]
data = data.append(dictSmall, ignore_index=True)
data = data.append(dictLarge, ignore_index=True)
data.drop_duplicates(subset='FoodName', inplace=True)
data[data['FoodName'].str.contains('assam prawn', case=False)]
```

<h4>Set labels for <strong>Cereals and cereal products</strong></h4>
<ul>
    <li>Update Vegan, Vegatarian and Halal</li>
    <li>Update Contains Beef</li>
</ul>

```python
NonVeganBiscuitsKeywords = ['chocolate','cream','ovaltine','cheese','butter']
VeganBiscuitsKeywords = ['cream cracker','peanut butter']

NonVeganBreadKeywords = ['beef','chocolate','butter','cheese','ham','chicken','milk','pork']
VeganBreadKeywords = ['butterscotch']
NonVegetarianBreadKeywords = ['beef', 'ham', 'chicken', 'pork']
BeefBreadKeywords = ['beef', 'ham', 'chicken', 'pork']

# Compile into a single list
NonVeganKeywords = np.unique(NonVeganBiscuitsKeywords+NonVeganBreadKeywords)
VeganKeywords = np.unique(VeganBiscuitsKeywords+VeganBreadKeywords)
NonVegetarianKeywords = np.unique(NonVegetarianBreadKeywords)
BeefKeywords = np.unique(BeefBreadKeywords)

# Update Vegan, Vegetarian and Halal
FoodGroups = 'CEREAL AND CEREAL PRODUCTS'
LabelNames = ['Vegan', 'Vegetarian', 'Halal']
value = False
# Non-Vegan
print(f'Update {LabelNames[0]} {value}:')
UpdateDataLabels(dataUpdate=data, foodGroups=FoodGroups, foodSubGroups=None, labelNames=LabelNames, keyWords=NonVeganKeywords,\
                 value=value)

# Vegan
value = True
print(f'Update {LabelNames[0]} {value}:')
UpdateDataLabels(dataUpdate=data, foodGroups=FoodGroups, foodSubGroups=None, labelNames=LabelNames, keyWords=VeganKeywords,\
                 value=value)

# Vegetarian
LabelNames = ['Vegetarian', 'Halal']
value = False
print(f'Update {LabelNames[0]} {value}:')
UpdateDataLabels(dataUpdate=data, foodGroups=FoodGroups, foodSubGroups=None, labelNames=LabelNames,\
                 keyWords=NonVegetarianKeywords, value=value)

# Contains Beef
LabelNames = ['ContainsBeef']
value = True
print(f'Update {LabelNames[0]} {value}:')
UpdateDataLabels(dataUpdate=data, foodGroups=FoodGroups, foodSubGroups=None, labelNames='ContainsBeef',\
                 keyWords=BeefKeywords, value=value)
```
